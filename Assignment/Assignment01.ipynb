**Que 5. Using a while loop, verify if the number A is purely divisible by number B and if so then how many times it can be divisible.**

```python
#Input dividend and divisor
A = int(input("Enter the dividend (A): "))
B = int(input("Enter the divisor (B): "))

#count division
divisions_count = 0

#While loop
while A % B == 0:
    A = A // B
    divisions_count += 1

#Display result
if divisions_count > 0:
    print(f"{A} is purely divisible by {B} {divisions_count} times.")
else:
    print(f"{A} is not purely divisible by {B}.")
```

**Que 7. What do you understand about mutable and immutable data types? Give examples for both showing this property.**

**Mutable Data Types:**

* Objects of mutable data types can be modified after creation.
*Changes to the object do not create a new object; instead, they modify the existing object in-place.
*Examples of mutable data types in Python include lists, dictionaries, and sets.


```python
# Example of a mutable list
list = [1, 2, 3, 4]
print("Original List:", list)

list.append(5)
print("Modified List:", list)
```

**Immutable Data Types:**

* Objects of immutable data types cannot be modified after creation.
* Any operation that seems to modify the object actually creates a new object.
* Examples of immutable data types in Python include integers, floats, strings, and tuples.


```python
# Example of an immutable tuple
tuple = (1, 2, 3, 4)
print("Original Tuple:", tuple)

# Create a new tuple with an additional element
new_tuple = tuple + (5,)
print("New Tuple:", new_tuple)
```
